```python
import pandas as pd
from IPython.display import Image
import pydotplus
from sklearn.externals.six import StringIO
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv('./titanic/data_clean1.csv')
titanic_df.head()
```

Erotellaan muuttuja "Survived" Erilleen muista

```python
X = titanic_df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_C', 'Embarked_Q', 'Embarked_S']]
y = titanic_df['Survived']
print(y.shape)
print(X.shape)
X
```

TehdÃ¤Ã¤n train-test split, kÃ¤ytetÃ¤Ã¤n 75% datasta opetukseen ja 25% testaukseen

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True, random_state=100)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
X_train
```

Seuraavaksi tehdÃ¤Ã¤n pÃ¤Ã¤tÃ¶spuu

```python
decision_tree_clf = DecisionTreeClassifier(max_depth=4, min_samples_leaf=20, random_state=1)
decision_tree_clf.fit(X_train, y_train)
```

Mitataan vielÃ¤, kuinka hyvin algoritmi pÃ¤Ã¤ttelee selviytymisen.

```python
decision_tree_clf.score(X_test, y_test)
```

TehdÃ¤Ã¤n vielÃ¤ neuroverkko

```python
neural_network_clf = MLPClassifier((32,2), learning_rate_init=0.001, solver='adam', batch_size=32, random_state=1)
neural_network_clf.fit(X_train, y_train)
```

Mitataan vielÃ¤ neuroverkon suoriutuminen

```python
neural_network_clf.score(X_test, y_test)
```
